```python
#----------------------- USUARIO ----------------------

from abc import ABC, abstractmethod

class Usuario(ABC):
    def __init__(self, nombre, id_usuario):
        self._nombre = nombre
        self._id_usuario = id_usuario

    @abstractmethod
    def solicitar_prestamo(self, prestamo):
        pass

#subclases
class EstudianteDeSecundaria(Usuario):
    def solicitar_prestamo(self, prestamo):
        return f"{self._nombre} ha solicitado el préstamo {prestamo._publicacion.obtener_informacion()} como estudiante de secundaria"

class EstudianteDePreparatoria(Usuario):
    def solicitar_prestamo(self, prestamo):
        return f"{self._nombre} ha solicitado el préstamo {prestamo._publicacion.obtener_informacion()} como estudiante de preparatoria"

class EstudianteDeUniversidad(Usuario):
    def solicitar_prestamo(self, prestamo):
        return f"{self._nombre} ha solicitado el préstamo {prestamo._publicacion.obtener_informacion()} como estudiante de universidad"

class EstudianteDePosgrado(Usuario):
    def solicitar_prestamo(self, prestamo):
        return f"{self._nombre} ha solicitado el préstamo {prestamo._publicacion.obtener_informacion()} como estudiante de posgrado"

class EstudianteDeIntercambio(Usuario):
    def solicitar_prestamo(self, prestamo):
        return f"{self._nombre} ha solicitado el préstamo {prestamo._publicacion.obtener_informacion()} como estudiante de intercambio"

class Investigador(Usuario):
    def solicitar_prestamo(self, prestamo):
        return f"{self._nombre} ha solicitado el préstamo {prestamo._publicacion.obtener_informacion()} como investigador"

class Administrativo(Usuario):
    def solicitar_prestamo(self, prestamo):
        return f"{self._nombre} ha solicitado el préstamo {prestamo._publicacion.obtener_informacion()} como administrativo"

class Bibliotecario(Usuario):
    def solicitar_prestamo(self, prestamo):
        return f"{self._nombre} ha solicitado el préstamo {prestamo._publicacion.obtener_informacion()} como bibliotecario"

class AdultoINAPAM(Usuario):
    def solicitar_prestamo(self, prestamo):
        return f"{self._nombre} ha solicitado el préstamo {prestamo._publicacion.obtener_informacion()} como adulto INAPAM"

class AdultoComun(Usuario):
    def solicitar_prestamo(self, prestamo):
        return f"{self._nombre} ha solicitado el préstamo {prestamo._publicacion.obtener_informacion()} como adulto común"



#----------------------- PUBLICACION --------------------

from abc import ABC, abstractmethod

class Publicacion(ABC):
    def __init__(self, titulo, autor):
        self._titulo = titulo
        self._autor = autor
        self._disponible = True

    @abstractmethod
    def consultar_disponibilidad(self):
        pass
    def prestar(self):
        self._disponible = False
    def devolver(self):
        self._disponible = True
        return f"La publicación '{self._titulo}' de {self._autor} ha sido devuelta"
    def obtener_informacion(self):
        return f"Título: {self._titulo}, Autor: {self._autor}"

class Libro(Publicacion):
    def __init__(self, titulo, autor):
        super().__init__(titulo, autor)
        self._disponible = True

    def prestar(self):
        super().prestar()
    def devolver(self):
        super().devolver()

    def consultar_disponibilidad(self):
        if self._disponible:
          return f"El libro '{self._titulo}' de {self._autor} está disponible"
        else:
          return f"El libro '{self._titulo}' de {self._autor} no está disponible"

class Revista(Publicacion):
    def __init__(self, titulo, autor):
        super().__init__(titulo, autor)

    def prestar(self):
        self._disponible = False
    def devolver(self):
        self._disponible = True
        return f"La revista '{self._titulo}' de {self._autor} ha sido devuelta"

    def consultar_disponibilidad(self):
        return f"La revista '{self._titulo}' de {self._autor} está disponible"

class Novela(Publicacion):
    def __init__(self, titulo, autor):
        super().__init__(titulo, autor)
    def consultar_disponibilidad(self):
        return f"La novela '{self._titulo}' de {self._autor} está disponible"

class Comic(Publicacion):
    def __init__(self, titulo, autor):
        super().__init__(titulo, autor)
    def consultar_disponibilidad(self):
        return f"El comic '{self._titulo}' de {self._autor} está disponible"

class Diccionario(Publicacion):
    def __init__(self, titulo, autor):
        super().__init__(titulo, autor)
    def consultar_disponibilidad(self):
        return f"El diccionario '{self._titulo}' de {self._autor} está disponible"

class Manual(Publicacion):
    def __init__(self, titulo, autor):
        super().__init__(titulo, autor)
    def consultar_disponibilidad(self):
        return f"El manual '{self._titulo}' de {self._autor} está disponible"

class Enciclopedia(Publicacion):
    def __init__(self, titulo, autor):
        super().__init__(titulo, autor)
    def consultar_disponibilidad(self):
        return f"La enciclopedia '{self._titulo}' de {self._autor} está disponible"

class Articulo(Publicacion):
    def __init__(self, titulo, autor):
        super().__init__(titulo, autor)
    def consultar_disponibilidad(self):
        return f"El artículo '{self._titulo}' de {self._autor} está disponible"

class Tesis(Publicacion):
    def __init__(self, titulo, autor):
        super().__init__(titulo, autor)
    def consultar_disponibilidad(self):
        return f"La tesis '{self._titulo}' de {self._autor} está disponible"

class Boletin(Publicacion):
    def __init__(self, titulo, autor):
        super().__init__(titulo, autor)
    def consultar_disponibilidad(self):
        return f"El boletín '{self._titulo}' de {self._autor} está disponible"

#----------------------- PRESTAMO -----------------------

class Prestamo(ABC):
    def __init__(self, usuario, publicacion):
        self._usuario = usuario
        self._publicacion = publicacion

    @abstractmethod
    def calcular_tarifa(self):
        pass

#subclases
class PrestamoDiario(Prestamo):
    def calcular_tarifa(self):
        return f"Tarifa para préstamo de un día ${10}"

class PrestamoSemanal(Prestamo):
    def calcular_tarifa(self):
        return f"Tarifa para préstamo de una semana ${25}"

class PrestamoQuincenal(Prestamo):
    def calcular_tarifa(self):
        return f"Tarifa para préstamo de 15 días ${35}"

class PrestamoMensual(Prestamo):
    def calcular_tarifa(self):
        return f"Tarifa para préstamo de un mes ${50}"

class PrestamoFinSemana(Prestamo):
    def calcular_tarifa(self):
        return f"Tarifa fin para préstamo de un fin de semana ${15}"

class PrestamoVacaciones(Prestamo):
    def calcular_tarifa(self):
        return f"Tarifa para préstamo de vacaciones ${70}"

class PrestamoEstacional(Prestamo):
    def calcular_tarifa(self):
        return f"Tarifa para préstamo de periodo estacional ${100}"

class PrestamoExtraordinario(Prestamo):
    def calcular_tarifa(self):
        return f"Tarifa para préstamo extraordinario ${150}"

class PrestamoEmergencias(Prestamo):
    def calcular_tarifa(self):
        return f"Tarifa para préstamo de emergencia ${170}"

class PrestamoLargoPlazo(Prestamo):
    def calcular_tarifa(self):
        return f"Tarifa para préstamo de largo plazo ${215}"

#----------------------- NOTIFICACION -------------------

class Notificacion(ABC):
    def __init__(self, mensaje, usuario):
        self._mensaje = mensaje
        self._usuario = usuario

    @abstractmethod
    def enviar_notificacion(self):
        pass

#subclases
class NotificacionCorreo(Notificacion):
    def enviar_notificacion(self):
        return f"Enviando notificación por correo electrónico a {self._usuario}: {self._mensaje}"

class NotificacionSMS(Notificacion):
    def enviar_notificacion(self):
        return f"Enviando notificación por SMS a {self._usuario}: {self._mensaje}"

class NotificacionWhatsApp(Notificacion):
    def enviar_notificacion(self):
        return f"Enviando notificación por WhatsApp a {self._usuario}: {self._mensaje}"

class NotificacionRedesSociales(Notificacion):
    def enviar_notificacion(self):
        return f"Enviando notificación por redes sociales a {self._usuario}: {self._mensaje}"

class NotificacionLlamada(Notificacion):
    def enviar_notificacion(self):
        return f"Realizando llamada a {self._usuario} con el mensaje: {self._mensaje}"

class NotificacionCarta(Notificacion):
    def enviar_notificacion(self):
        return f"Enviando carta a {self._usuario} con el mensaje: {self._mensaje}"

class NotificacionTablonAnuncios(Notificacion):
    def enviar_notificacion(self):
        return f"Publicando anuncio en el tablón de anuncios para {self._usuario}: {self._mensaje}"

class NotificacionMensajeria(Notificacion):
    def enviar_notificacion(self):
        return f"Enviando mensaje a {self._usuario} por mensajería instantánea: {self._mensaje}"

class NotificacionSistemaPrestamos(Notificacion):
    def enviar_notificacion(self):
        return f"Enviando notificación por sistema de préstamos a {self._usuario}: {self._mensaje}"

class NotificacionMensajeVoz(Notificacion):
    def enviar_notificacion(self):
        return f"Dejando mensaje de voz para {self._usuario} con el mensaje: {self._mensaje}"


#----------------------- EVENTO -------------------------

class Evento(ABC):
    def __init__(self, nombre, detalles):
        self._nombre = nombre
        self._detalles = detalles

    @abstractmethod
    def registrar_asistencia(self, usuario):
        pass
#subclases
class EventoDeLanzamientoDeLibro(Evento):
    def registrar_asistencia(self, usuario):
        return f"{usuario} ha registrado su asistencia en el evento de lanzamiento de libro '{self._nombre}'"

class EventoDeFirmaDeLibros(Evento):
    def registrar_asistencia(self, usuario):
        return f"{usuario} ha registrado su asistencia en el evento de firma de libros '{self._nombre}'"

class EventoDeLectura(Evento):
    def registrar_asistencia(self, usuario):
        return f"{usuario} ha registrado su asistencia en el evento de lectura '{self._nombre}'"

class EventoDeClubDeLectura(Evento):
    def registrar_asistencia(self, usuario):
        return f"{usuario} ha registrado su asistencia en el evento de club de lectura '{self._nombre}'"

class EventoDeDescuentos(Evento):
    def registrar_asistencia(self, usuario):
        return f"{usuario} ha registrado su asistencia en el evento de descuentos '{self._nombre}'"

class EventoDeTallerDeEscritura(Evento):
    def registrar_asistencia(self, usuario):
        return f"{usuario} ha registrado su asistencia en el evento de taller de escritura '{self._nombre}'"

class EventoDePresentacionDeLibro(Evento):
    def registrar_asistencia(self, usuario):
        return f"{usuario} ha registrado su asistencia en el evento de presentación de libro '{self._nombre}'"

class EventoDeFeriaDelLibro(Evento):
    def registrar_asistencia(self, usuario):
        return f"{usuario} ha registrado su asistencia en el evento de feria del libro '{self._nombre}'"

class EventoDeLecturaDeCuentosInfantiles(Evento):
    def registrar_asistencia(self, usuario):
        return f"{usuario} ha registrado su asistencia en el evento de lectura de cuentos infantiles '{self._nombre}'"

class EventoDeEncuentroConElAutor(Evento):
    def registrar_asistencia(self, usuario):
        return f"{usuario} ha registrado su asistencia en el evento de encuentro con el autor '{self._nombre}'"


#----------------------- BIBLIOTECA -----------------------
class Biblioteca:
    def __init__(self):
        self.usuarios = []
        self.publicaciones = []
        self.prestamos = []
        self.notificaciones = []
        self.eventos = []

    def agregar_usuario(self, usuario):
        self.usuarios.append(usuario)

    def agregar_publicacion(self, publicacion):
        self.publicaciones.append(publicacion)

    def agregar_prestamo(self, prestamo):
        self.prestamos.append(prestamo)

    def agregar_notificacion(self, notificacion):
        self.notificaciones.append(notificacion)

    def agregar_evento(self, evento):
        self.eventos.append(evento)

    def solicitar_prestamo(self, usuario, publicacion):
        prestamo = PrestamoDiario(usuario, publicacion)
        self.agregar_prestamo(prestamo)
        return f"Se ha solicitado el préstamo de {publicacion.obtener_informacion()} para {usuario._nombre}"

    def enviar_notificacion(self, notificacion):
        return notificacion.enviar_notificacion()

    def registrar_asistencia(self, evento, usuario):
        return evento.registrar_asistencia(usuario._nombre)

    def consultar_disponibilidad(self, publicacion):
        return publicacion.consultar_disponibilidad()

    def calcular_tarifa(self, prestamo):
        return prestamo.calcular_tarifa()

    def mostrar_menu(self):
        print("\nBienvenido/a a la Biblioteca Virtual S&C.")
        print("\n1. Agregar usuario")
        print("2. Agregar publicación")
        print("3. Solicitar préstamo")
        print("4. Enviar notificación")
        print("5. Registrar asistencia a evento")
        print("6. Consultar disponibilidad de publicación")
        print("7. Devolver publicacion")
        print("8. Salir")

    def ejecutar_accion(self, opcion):
      #----------------- Menu para escoger usuario---------------------
        if opcion == 1:
            nombre = input("\nIngrese el nombre del usuario: ")
            id_usuario = input("Ingrese el ID del usuario: ")
            print("Seleccione el tipo de usuario que desea agregar:")
            print("1. Estudiante de secundaria")
            print("2. Estudiante de preparatoria")
            print("3. Estudiante de universidad")
            print("4. Estudiante de posgrado")
            print("5. Estudiante de intercambio")
            print("6. Investigador")
            print("7. Administrativo")
            print("8. Bibliotecario")
            print("9. Adulto INAPAM")
            print("10. Adulto común")
            tipo_usuario = int(input("Ingrese la opción que desea seleccionar: "))
            if tipo_usuario == 1:
                usuario = EstudianteDeSecundaria(nombre, id_usuario)
            elif tipo_usuario == 2:
                usuario = EstudianteDePreparatoria(nombre, id_usuario)
            elif tipo_usuario == 3:
                usuario = EstudianteDeUniversidad(nombre, id_usuario)
            elif tipo_usuario == 4:
                usuario = EstudianteDePosgrado(nombre, id_usuario)
            elif tipo_usuario == 5:
                usuario = EstudianteDeIntercambio(nombre, id_usuario)
            elif tipo_usuario == 6:
                usuario = Investigador(nombre, id_usuario)
            elif tipo_usuario == 7:
                usuario = Administrativo(nombre, id_usuario)
            elif tipo_usuario == 8:
                usuario = Bibliotecario(nombre, id_usuario)
            elif tipo_usuario == 9:
                usuario = AdultoINAPAM(nombre, id_usuario)
            elif tipo_usuario == 10:
                usuario = AdultoComun(nombre, id_usuario)
            biblioteca.agregar_usuario(usuario)
            print("Usuario agregado con éxito")
            return True

        #------------------- menu para agregar publicacion --------------------
        elif opcion == 2:
            titulo = input("\nIngrese el título de la publicación: ")
            autor = input("Ingrese el autor de la publicación: ")
            print("Seleccione el tipo de publicación que desea agregar:")
            print("1. Libro")
            print("2. Revista")
            print("3. Novela")
            print("4. Comic")
            print("5. Diccionario")
            print("6. Manual")
            print("7. Tesis")
            print("8. Enciclopedia")
            print("9. Artículo")
            print("10. Boletín")
            tipo_publicacion = int(input("Ingrese la opción que desea seleccionar: "))
            if tipo_publicacion == 1:
                publicacion = Libro(titulo, autor)
            elif tipo_publicacion == 2:
                publicacion = Revista(titulo, autor)
            elif tipo_publicacion == 3:
                publicacion = Novela(titulo, autor)
            elif tipo_publicacion == 4:
                publicacion = Comic(titulo, autor)
            elif tipo_publicacion == 5:
                publicacion = Diccionario(titulo, autor)
            elif tipo_publicacion == 6:
                publicacion = Manual(titulo, autor)
            elif tipo_publicacion == 7:
                publicacion = Tesis(titulo, autor)
            elif tipo_publicacion == 8:
                publicacion = Enciclopedia(titulo, autor)
            elif tipo_publicacion == 9:
                publicacion = Articulo(titulo, autor)
            elif tipo_publicacion == 10:
                publicacion = Boletin(titulo, autor)
            biblioteca.agregar_publicacion(publicacion)
            print("Publicación agregada con éxito")
            return True

        # ------------------- menu para solicitar prestamo -----------------------
        elif opcion == 3:
            nombre_usuario = input("\nIngrese el nombre del usuario que solicita el préstamo: ")
            titulo_publicacion = input("Ingrese el título de la publicación que se solicita: ")
            usuario = next((u for u in self.usuarios if u._nombre == nombre_usuario), None)
            publicacion = next((p for p in self.publicaciones if p._titulo == titulo_publicacion), None)
            if usuario and publicacion:
                if publicacion._disponible:
                    publicacion.prestar()
                    print(biblioteca.solicitar_prestamo(usuario, publicacion))
                    print("La publicacion está disponible para préstamo")
                    print("Seleccione el tiempo para el préstamo:")
                    print("1. Un día")
                    print("2. Una semana")
                    print("3. 15 días")
                    print("4. Un mes")
                    print("5. Fin de semana")
                    print("6. Por vacaciones")
                    print("7. Estacional")
                    print("8. Extraordinario")
                    print("9. Emergencias")
                    print("10. De largo plazo")

                    tipo_prestamo = int(input("Ingrese la opción que desea seleccionar: "))

                    if tipo_prestamo == 1:
                        prestamo = PrestamoDiario(usuario, publicacion)
                    elif tipo_prestamo == 2:
                        prestamo = PrestamoSemanal(usuario, publicacion)
                    elif tipo_prestamo == 3:
                        prestamo = PrestamoQuincenal(usuario, publicacion)
                    elif tipo_prestamo == 4:
                        prestamo = PrestamoMensual(usuario, publicacion)
                    elif tipo_prestamo == 5:
                        prestamo = PrestamoFinSemana(usuario, publicacion)
                    elif tipo_prestamo == 6:
                        prestamo = PrestamoVacaciones(usuario, publicacion)
                    elif tipo_prestamo == 7:
                        prestamo = PrestamoEstacional(usuario, publicacion)
                    elif tipo_prestamo == 8:
                        prestamo = PrestamoExtraordinario(usuario, publicacion)
                    elif tipo_prestamo == 9:
                        prestamo = PrestamoEmergencias(usuario, publicacion)
                    elif tipo_prestamo == 10:
                        prestamo = PrestamoLargoPlazo(usuario, publicacion)
                    else:
                        print("Opción no válida")
                        True

                    tarifa = biblioteca.calcular_tarifa(prestamo)     # tarifa de acuerdo al prestamo
                    print(tarifa)

                else:
                    print("La publicación no está disponible")

            else:
                print("Usuario o publicación no encontrados")

        # ----------------------- menu para notificaciones ----------------------
        elif opcion == 4:
            mensaje = input("\nIngrese el mensaje de la notificación: ")
            nombre_usuario = input("Ingrese el nombre del usuario que recibirá la notificación: ")
            usuario = next((u for u in self.usuarios if u._nombre == nombre_usuario), None)
            if usuario:
                print("Seleccion por qué medio desea enviar la notificación:")
                print("1. Correo")
                print("2. SMS")
                print("3. WhatsApp")
                print("4. Redes Sociales")
                print("5. Llamada")
                print("6. Carta")
                print("7. Tablon de anuncios")
                print("8. Mensajería instantánea")
                print("9. Sistema de préstamos")
                print("10. Mensaje de voz")
                tipo_notificacion = int(input("Ingrese la opción que desea seleccionar: "))
                if tipo_notificacion == 1:
                    notificacion = NotificacionCorreo(mensaje, usuario._nombre)
                if tipo_notificacion == 2:
                    notificacion = NotificacionSMS(mensaje, usuario._nombre)
                if tipo_notificacion == 3:
                    notificacion = NotificacionWhatsApp(mensaje, usuario._nombre)
                if tipo_notificacion == 4:
                    notificacion = NotificacionRedesSociales(mensaje, usuario._nombre)
                if tipo_notificacion == 5:
                    notificacion = NotificacionLlamada(mensaje, usuario._nombre)
                if tipo_notificacion == 6:
                    notificacion = NotificacionCarta(mensaje, usuario._nombre)
                if tipo_notificacion == 7:
                    notificacion = NotificacionTablonAnuncios(mensaje, usuario._nombre)
                if tipo_notificacion == 8:
                    notificacion = NotificacionMensajeria(mensaje, usuario._nombre)
                if tipo_notificacion == 9:
                    notificacion = NotificacionSistemaPrestamos(mensaje, usuario._nombre)
                if tipo_notificacion == 10:
                    notificacion = NotificacionMensajeVoz(mensaje, usuario._nombre)
                print(self.enviar_notificacion(notificacion))
            else:
                print("Usuario no encontrado")

        # -------------------- para escoger evento -----------------------
        elif opcion == 5:
            print("\nEventos disponibles:")
            for i, evento in enumerate(self.eventos, start=1):
                print(f"{i}. {evento._nombre} - {evento._detalles}")

            opcion_evento = int(input("Ingrese el número del evento al que desea asistir: "))
            evento_seleccionado = self.eventos[opcion_evento - 1]

            nombre_usuario = input("Ingrese el nombre del usuario que asistirá al evento: ")
            usuario = next((u for u in self.usuarios if u._nombre == nombre_usuario), None)

            if usuario:
                print(self.registrar_asistencia(evento_seleccionado, usuario))
            else:
                print("Usuario no encontrado")

        # ---------------------------- para consulta de publicaiones --------------------
        elif opcion == 6:
            titulo_publicacion = input("\nIngrese el título de la publicación que se desea consultar: ")
            publicacion = next((p for p in self.publicaciones if p._titulo == titulo_publicacion), None)
            if publicacion:
                print(self.consultar_disponibilidad(publicacion))
            else:
                print("Publicación no encontrada")

        # ---------------------------- para devolver publicacion ------------------------
        elif opcion == 7:
            titulo_publicacion = input("\nIngrese el título de la publicación que se devolverá: ")
            publicacion = next((p for p in biblioteca.publicaciones if p._titulo == titulo_publicacion), None)
            if publicacion:
                publicacion.devolver()
                print("La publicación ha sido devuelta")
            else:
                print("Publicación no encontrada")
                return True

        # ---------------------- para salir ----------------------------------
        elif opcion == 8:
            print("\n.Gracias por utilizar la biblioteca virtual.")
            return False
        else:
            print("Opción no válida")
        return True

    def instanciasDefault(self):
        # Instancias de diferentes tipos de usuarios
        usuarios = [
            EstudianteDeSecundaria("Juan Pérez", 1),
            Investigador("Ana Gómez", 2),
            AdultoINAPAM("Roberto Hernández", 3),
            EstudianteDeUniversidad("Sofía Ramírez", 4)
        ]

        # Instancias de diferentes tipos de publicaciones
        publicaciones = [
            Libro("1984", "George Orwell"),
            Revista("National Geographic", "Variado"),
            Comic("Spider-Man", "Stan Lee"),
            Tesis("Aplicación de Machine Learning", "Carlos Ruiz"),
            Enciclopedia("Enciclopedia Británica", "Varios Autores")
        ]

        # Instancias de diferentes tipos de eventos
        eventos = [
            EventoDeLanzamientoDeLibro("Lanzamiento de 'El Quijote'", "Presentación del libro"),
            EventoDeClubDeLectura("Club de lectura de 'Cien Años de Soledad'", "Discusión del libro"),
            EventoDeFeriaDelLibro("Feria Internacional del Libro", "Venta de libros a precios especiales"),
            EventoDeTallerDeEscritura("Taller de Escritura Creativa", "Aprender a escribir ficción")
        ]

        #agregar instancias
        for usuario in usuarios:
            self.agregar_usuario(usuario)

        for publicacion in publicaciones:
            self.agregar_publicacion(publicacion)

        for evento in eventos:
            self.agregar_evento(evento)

    def ejecutar(self):
        while True:
            self.mostrar_menu()
            opcion = int(input("\nIngrese la opción que desea realizar: "))
            if not self.ejecutar_accion(opcion):
                break

biblioteca = Biblioteca()
biblioteca.instanciasDefault()
biblioteca.ejecutar()
```
